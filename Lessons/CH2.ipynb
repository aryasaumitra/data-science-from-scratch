# Visualizing Data

```python
from matplotlib import pyplot as plt
from collections import Counter
```

```python
years = [1950,1960,1970,1980,1990,2000,2010]
gdp = [300.2,543.3,1075.9,2862.69,5979.6,10289.7,14958.3]
```

```python
plt.plot(years,gdp,color='green',marker = 'o',linestyle='solid')
plt.title("Nominal GDP")
plt.ylabel("Billions of $")
plt.xlabel("Years")
plt.show()
```

## Bar Chart

```python
movies = ["Annie Hall", "Ben-Hur", "Casablanca", "Gandhi", "West Side Story"]
num_oscars = [5, 11, 3, 8, 10]
```

```python
plt.bar(range(len(movies)),num_oscars)
plt.title("Movie Awards")
plt.ylabel("# of Academy Awards")
plt.xticks(range(len(movies)),movies) #Maps movies to x-axis Indices at the cente of the bar
plt.show()
```

```python
xs = [i + 0.1 for i, _ in enumerate(movies)]

plt.bar(xs,num_oscars)
plt.title("Movie Awards")
plt.ylabel("# of Academy Awards")
plt.xticks([i + 0.5 for i, _ in enumerate(movies)],movies) # label x-axis with movie names at bar ends
plt.show()
```

## Histogram

```python
grades = [83,95,91,87,70,0,85,82,100,67,73,77,0]

histogram = Counter(min(grade //10*10,90) for grade in grades) #Bucket 100 with the 90
```

```python
histogram
```

```python
decile = lambda grade: grade // 10*10

histogram2 = Counter(decile(grade) for grade in grades) #Bucket 100 as it is
```

```python
histogram2
```

```python
plt.bar([x+5 for x in histogram.keys()],histogram.values(),10,edgecolor=(0,0,0)) #Since Bars Position in the Middle of an Interval we shift them by 5
plt.axis([-5,105,0,5]) # Careful while Moving the axis
plt.xticks([10*i for i in range(11)])
plt.xlabel("Decile")
plt.ylabel("# of Students")
plt.title("Distribution of Student Grades")
```

### Misleading Axis

```python
mentions = [500,505]
years = [2017,2018]
```

```python
plt.bar(years,mentions,0.8)
plt.xticks(years)
plt.ylabel("# Mention Count")
plt.ticklabel_format(useOffset=False)
plt.axis([2016.5,2018.5,499,506])
plt.title("Misleading Axis")
plt.show()
```

```python
plt.bar(years,mentions,0.8)
plt.xticks(years)
plt.ylabel("# Mention Count")
plt.ticklabel_format(useOffset=False)
plt.axis([2016.5,2018.5,0,550])
plt.title("Non - Misleading Axis")
plt.show()
```

## Line Chart

```python
variance     = [1,2,4,8,16,32,64,128,256]
bias_squared = [256,128,64,32,16,8,4,2,1]
total_error  = [x + y for x, y in zip(variance, bias_squared)]
```

```python
xs = [i for i,_ in enumerate(variance)]
```

```python
plt.plot(xs,variance, 'g-',label='variance') #Green Solid Line
plt.plot(xs,bias_squared, 'r-.',label='bias^2') #Red Dot-Dashed Line
plt.plot(xs,total_error, 'b:',label='total error') #Blue Dotted

plt.legend(loc=9) #9 means Top Center. How ?
plt.xlabel("Model Complexity")
plt.xticks([]) # No Index in X-Axis
plt.title("Bias-Variance Tradeoff")
plt.show()
```

## ScatterPlots

```python
friends = [ 70, 65, 72, 63, 71, 64, 60, 64, 67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']
```

```python
plt.scatter(friends,minutes)
```

```python
# Mark and Label Points
plt.scatter(friends,minutes)
for label,friend_count,minute_count in zip(labels,friends,minutes):
    plt.annotate(label,
                xy = (friend_count,minute_count),
                xytext = (5,-5),
                textcoords = 'offset points'
               )
plt.title("Daily Minutes vs, Number of Friends")
plt.xlabel("# of Friends")
plt.ylabel("Daily Minutes on Site")
plt.show()
```

### ScatterPlot with undefined Axis

```python
test_1_grades = [ 99, 90, 85, 97, 80]
test_2_grades = [100, 85, 60, 90, 70]
```

```python
plt.scatter(test_1_grades, test_2_grades)
plt.title("Axes Aren't Comparable")
plt.xlabel("test 1 grade")
plt.ylabel("test 2 grade")
plt.show()
```

### ScatterPlot with defined Axis

```python
plt.scatter(test_1_grades, test_2_grades)
plt.axis("equal")
plt.title("Axes Are Comparable")
plt.xlabel("test 1 grade")
plt.ylabel("test 2 grade")
plt.show()
```
